```python
import requests
import io
import pandas as pd

```

```python
url = "https://www.ncei.noaa.gov/oa/local-climatological-data/v2/doc/lcdv2-station-list.txt"
resp = requests.get(url)
resp.raise_for_status()
text = resp.text 
f = io.StringIO(text)
f
```

```python

def parse_line(line):
    parts = line.strip().split()
    if len(parts) < 5:
        return None
    sid = parts[0]
    lat = float(parts[1])
    lon = float(parts[2])
    elev = float(parts[3])
    name = " ".join(parts[4:])
    return (sid, lat, lon, elev, name)

records = []
for ln in f:
    parsed = parse_line(ln)
    if parsed:
        records.append(parsed)

df = pd.DataFrame(records, columns=["station_id", "latitude", "longitude", "elevation", "station_name"])

df.head()
```

```python
import pandas as pd
flight = pd.read_csv("T_ONTIME_REPORTING 3.csv",dtype={'DEP_TIME': str})
flight = flight[flight['ORIGIN_CITY_NAME'] == 'Raleigh/Durham, NC']
flight
```

```python
flight.head()
```

```python
weather = pd.read_csv("clean_weather_RDU_2022_2024.csv")
weather
```

```python
flight["FL_DATE"] = pd.to_datetime(flight["FL_DATE"], format="%m/%d/%Y %I:%M:%S %p")
weather["DATE"] = pd.to_datetime(weather["DATE"])
flight = flight.sort_values("FL_DATE")
weather = weather.sort_values("DATE")
print(flight)
print(weather)
```

```python
flight['DEP_TIME'] = flight['DEP_TIME'].astype(str).str.strip()
valid_mask = flight['DEP_TIME'].str.fullmatch(r'\d{4}', na=False)
invalid_rows = flight[~valid_mask]
invalid_count = len(invalid_rows)
invalid_ratio = invalid_count / len(flight) * 100
print(invalid_rows)
print(invalid_count)
print(invalid_ratio)
```

```python
print(flight['DEP_TIME'])
```

```python
flight = flight[valid_mask].copy()
flight = flight.reset_index(drop=True)
```

```python
flight
```

```python
flight['DEP_TIME'] = flight['DEP_TIME'].astype(str).str.zfill(4)
flight['DEP_HOUR'] = flight['DEP_TIME'].str[:2].astype(int)
flight['DEP_MIN'] = flight['DEP_TIME'].str[2:].astype(int)
flight['DEP_DATETIME'] = flight.apply(
    lambda row: row['FL_DATE'] + pd.Timedelta(hours=row['DEP_HOUR'], minutes=row['DEP_MIN']),
    axis=1
)
```

```python
flight
```

```python
flight = flight.sort_values('DEP_DATETIME')
weather = weather.sort_values('DATE')
flight
```

```python
merged = pd.merge_asof(
    flight,
    weather,
    left_on="DEP_DATETIME",
    right_on="DATE",
    direction="nearest",   
    tolerance=pd.Timedelta("2H")  
)
```

```python
merged
```

```python
print(merged["DATE"].isna().mean()) 
```

```python
merged["time_diff_min"] = (merged["DEP_DATETIME"] - merged["DATE"]).abs().dt.total_seconds() / 60
```

```python
print(merged)
```

```python
nan_counts = merged.isna().sum()
print(nan_counts)
```

```python
total_weather_delay = merged['WEATHER_DELAY'].sum()
total_weather_delay
```

```python
merged.to_csv("rdu_flight_weather.csv", index=False)
```

```python
from pathlib import Path
root_path = Path("processed_flight_data")
csv_files = list(root_path.rglob("*.csv")) 
print(len(csv_files))
df_list = []
for file in csv_files:
    try:
        df = pd.read_csv(file, dtype=str)  
        df["source_file"] = file.name   
        print(file.name)
        df_list.append(df)
    except Exception as e:
        print(f"⚠️ failure load: {file}，for: {e}")
data = pd.concat(df_list, ignore_index=True)
data
```

```python
data['FL_DATE'] = pd.to_datetime(data['FL_DATE'], format='%Y-%m-%d')
data = data.sort_values('FL_DATE')
```
